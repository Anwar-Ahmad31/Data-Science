8. Reverse a vector (first element becomes last) (★☆☆)
   

```python
arr = np.arange(10,20)
print(arr)
print(arr[ : : -1])
```

## Learnings

1. Reverse ka matlab array ko ulta karna hota hai.
2. Reverse karne ke baad:
   - First element → Last
   - Last element → First
3. NumPy me slicing ka use karke vector ko reverse kar sakte hain.
4. Slicing ka syntax hota hai:
   
   arr[start:stop:step]

5. Agar step = -1 ho to array reverse direction me traverse hota hai.
6. `arr[::-1]` poore array ko reverse kar deta hai.  


    ## Objective

- Array Slicing ko samajhna.
- Negative step (-1) ka use samajhna.
- Reverse traversal ka concept seekhna.


    ## Common Mistake

Maine pehle socha tha ki reverse ka matlab mirror banana hota hai.

Example:
[1,2,3] → [1,2,3,3,2,1] ❌

Lekin reverse ka matlab hota hai:

[1,2,3] → [3,2,1] ✅ 


    ## Revision Point

Reverse = arr[::-1]

10. Find indices of non-zero elements from [1,2,0,0,4,0] (★☆☆)

```python
arr =np.array([1,2,0,0,4,0])

```

```python
arr !=0
```

## Learnings

1. `np.where()` condition ke basis par indices return karta hai.
2. Pehle condition likhte hain, jaise:
   - arr != 0
   - arr > 10
   - arr == 5
3. Jahan condition `True` hoti hai, unka index return hota hai.
4. Jahan condition `False` hoti hai, unhe ignore kar diya jata hai.
