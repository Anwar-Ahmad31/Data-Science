41. How to sum a small array faster than np.sum? (★★☆)



````python
arr= np.array([1,2,8,4])
sum(arr)
````

### Learning:

1. `np.sum()` is optimized for large NumPy arrays.

2. For very small arrays, Python built-in functions can sometimes be faster because NumPy function calls have overhead.

3. Performance depends on data size, not only on the function.

4. Always choose tools according to the problem size.

````python
import numpy as np
import time

# small array
arr = np.random.randint(1, 100, size=10)

print("Array:", arr)

# Python built-in sum()
start = time.time()

result1 = sum(arr)

end = time.time()

print("Python sum result:", result1)
print("Python sum time:", end - start)


# NumPy np.sum()
start = time.time()

result2 = np.sum(arr)

end = time.time()

print("NumPy sum result:", result2)
print("NumPy sum time:", end - start)
````
